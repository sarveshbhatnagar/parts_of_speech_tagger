```python
test_file = "POS.test"
```

```python
with open(test_file, "r") as f:
    data = f.read()
```

```python
data = data.split("\n")
```

```python
data
```

```python
import re
```

```python
pattern = "(.*)\/([A-Z,.\$:#)('`\.,\|]+)"
```

```python
def find_tags( rawword):
        """
        Finds the word,tags in the raw word

        :param rawword: raw word
        :return: word,tags
        """
        pattern = "(.*)\/([A-Z,.\$:#)('`\.,\|]+)"
        try:
            return re.findall(pattern, rawword)[0]
        except:
            # In case parsing fails
            return "SOMEWORD", "NN"
```

```python
z = []
for i in range(len(data)):
    z.append(data[i].split(" "))
```

```python
find_tags(z[0][0])
```

```python
a = []
for lis in z:
    for rawword in lis:
        a.append(find_tags(rawword))
```

```python
a
```

```python
unzipped_object = zip(*a)
unzipped_list = list(unzipped_object)
```
